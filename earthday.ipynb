# ***Industrial Waste***

#### ***Creating a Sub Dataset from the main dataset***

```python
data=df[['country_name','gdp','population_population_number_of_people','special_waste_industrial_waste_tons_year']]
```

```python
data.head()
```

```python
data.info()
```

```python
data=data.dropna()
```

```python
data.tail()
```

```python
data.isnull().sum()
```

#### ***Sorting the values in descending order*** 

```python
data1 = data.sort_values('special_waste_industrial_waste_tons_year', ascending = False) 
```

```python
data1.head()
```

#### ***Visualizing which country has the highest Industrial Waste Ton/Year on earth*** 

```python
fig = plt.figure(figsize = (10,5))
plt.title('Industrial Waste Tons/Year',size=20)
plt.xlabel('Country',size=15)
plt.ylabel('Industrial Waste in Tons',size=15)
plt.xticks(rotation=90)
plt.bar(data1['country_name'][:20],data1['special_waste_industrial_waste_tons_year'].values[:20],color='purple')
```

## ***OBERSERVATION:***

##  *US , China , Canada , Japan are the few countries which has highest tons of Industrial Waste in a Year*
